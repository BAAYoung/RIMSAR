```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
dx = 0.0001
dp = 0.025
edge_factor = 0.75
x = np.arange(0.3-dp*edge_factor,0.7+dx+dp*edge_factor,dx)




particle_coords = np.arange(0.3,0.7+dp,dp)

#particle_coords = np.random.rand(25)

plt.figure(figsize=(16,16))



plt.subplot(2,2,1)
sigma = 0.001
density = x*0
for i in range(0,particle_coords.shape[0]):
    density += np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
plt.plot(x,density)

plt.subplot(2,2,2)
sigma = 0.01
density = x*0
for i in range(0,particle_coords.shape[0]):
    density += np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
plt.plot(x,density)

plt.ylim([0,50])


plt.subplot(2,2,3)
sigma = 0.025
density = x*0
for i in range(0,particle_coords.shape[0]):
    density += np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
plt.plot(x,density)
plt.ylim([0,50])

plt.subplot(2,2,4)
#sigma = 0.075
sigma = 0.1
density = x*0
for i in range(0,particle_coords.shape[0]):
    density += np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
plt.plot(x,density)
plt.ylim([0,50])
```

```python
#with renormalisation
plt.figure(figsize=(16,16))
plt.subplot(2,2,1)
sigma = 0.001
density = x*0
for i in range(0,particle_coords.shape[0]):
    sol = np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
    sol /= np.trapz(sol)*dx
    density += sol
plt.plot(x,density)

plt.subplot(2,2,2)
sigma = 0.01
density = x*0
for i in range(0,particle_coords.shape[0]):
    sol = np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
    sol /= np.trapz(sol)*dx
    density += sol 
plt.plot(x,density)

plt.ylim([0,50])

plt.subplot(2,2,3)
sigma = 0.025
density = x*0
for i in range(0,particle_coords.shape[0]):
    sol = np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
    sol /= np.trapz(sol)*dx
    density += sol
plt.plot(x,density)
plt.ylim([0,50])

plt.subplot(2,2,4)
#sigma = 0.075
sigma = 10
density = x*0
for i in range(0,particle_coords.shape[0]):
    sol = np.exp(-0.5*(x-particle_coords[i])**2/sigma**2    )/(sigma*np.sqrt(2*3.14159))
    sol /= np.trapz(sol)*dx
    density += sol
plt.plot(x,density)
plt.ylim([0,50])
```
