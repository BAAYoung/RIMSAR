```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
dx = 5.0
xbi, zbi = np.meshgrid(np.arange(0.0,1024,dx),np.arange(0.0,1024,dx))


pixel2real = 0.0001848337
xbi *= pixel2real
zbi *= -pixel2real

u = xbi*0
v = xbi*0
rho = xbi*0
for j in range(0,5000):
    density = np.loadtxt('RCG_c0_01/rho'+str(j)+'.dat')
    momentumx = np.loadtxt('RCG_c0_01/momentumx'+str(j)+'.dat')
    momentumz = np.loadtxt('RCG_c0_01/momentumz'+str(j)+'.dat')
    u += momentumx/density
    v += momentumz/density
    rho += density 
```

```python
j = 10
plt.contourf(xbi,zbi,np.sqrt(u**2 + v**2),levels=100)
```
