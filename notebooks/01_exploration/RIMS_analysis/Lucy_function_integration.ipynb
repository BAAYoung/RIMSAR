```python
from sympy import * 
```

```python
xc, zc, x2, x1, y2, y1, Delta_r, theta, c, xp, zp, r, R, theta1, theta2 = symbols('x_c z_c x_2 x_1 y_2 y_1 Delta_r theta c x_p, z_p r R theta_1 theta_2')
```

```python
W = -3*(Delta_r/c)**4 + 8*(Delta_r/c)**3 -6*(Delta_r/c)**2 + 1
W = Piecewise(
    (0, Delta_r > c),
    (-3*(Delta_r/c)**4 + 8*(Delta_r/c)**3 -6*(Delta_r/c)**2 + 1, True)
)



plot(W.subs(c,1),(Delta_r,0,1.5))
```

```python
#integrating over the portioned circle:
Delta_r = simplify(sqrt(  (r*cos(theta) - xp)**2  +  (r*sin(theta) - zp)**2      ))
Delta_r = sqrt(xp**2 + zp**2 + r**2 - 2*xp*r*cos(theta) - 2*zp*r*sin(theta))

W = Piecewise(
    (0, Delta_r > c),
    (-3*(Delta_r/c)**4 + 8*(Delta_r/c)**3 -6*(Delta_r/c)**2 + 1, True)
)

pprint(Delta_r)
```

```python
#Integrating:

disk_sector = integrate(integrate(W*r,(r,0,R)), (theta, theta1, theta2))
```
