```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dx = 0.2
x,z = np.meshgrid(np.arange(-2,2+dx,dx),np.arange(-2,0+dx,dx))
r = np.maximum(np.abs(x),np.abs(z))
theta = np.arctan2(z,x)
xc = r*np.cos(theta)
zc = r*np.sin(theta)
plt.plot(xc.flatten(),zc.flatten(),'.')
```

```python
#checking no point is duplicated:
N = xc.flatten().shape[0]
min_dist = 1
xc = xc.flatten()
zc = zc.flatten()
for i in range(0,N):
    for j in range(0,N):
        if i!=j:
            r_ij = (xc[i] - xc[j])**2 + (zc[i] - zc[j])**2
            min_dist = min(r_ij,min_dist)

print(min_dist)
```
