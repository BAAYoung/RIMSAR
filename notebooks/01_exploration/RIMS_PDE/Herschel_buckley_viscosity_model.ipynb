```python
tau_0, z, k, mu_0, dudz, n, gamma_0, g, alpha, h, c_0 = symbols('tau_0 z k mu_0 dudz n gamma_0 g alpha h c_0')

#defining viscosity:

test_pw = Piecewise((0,z<-1),(z**2, z<=1),(z, True))
mu_flow = k*dudz**(n-1) + tau_0*dudz**(-1)
#tau_0 = 1
#mu_eff = Piecewise((0,dudz<gamma_0),(mu_flow, True))
alpha = 3.14159/4
g = 1
h = 1
n = 2
k = 1
```

```python
A = (g*sin(alpha)/k)**(1/n)
u = A*n*(h**(1/n + 1) - (h-z)**(1/n + 1))/(n+1)
mu_flow = k*u.diff(z)**(n-1) + tau_0*u.diff(z)**(-1)
plot(u,(z,0,h))
pprint(u)
pprint(u.diff(z).subs(z,h))
```
